```python
df2 = df.sort_values(by=['Year', 'Qtr', 'Total Offtake'], ascending=[True, True, False])
df2['Rank'] = df.groupby(['Year', 'Qtr'])['Total Offtake'].rank(ascending=False, method='dense')
```

```python
def assign_slab(rank):
    if rank <= 25 :
        return "Top"
    elif rank <= 50:
        return "High"
    elif rank <= 500:
        return "Medium"
    else:
        return "Low"
```

```python
df2['Rank Slab'] = df2['Rank'].apply(assign_slab)
```
